```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
dbEffectNoSkus = pd.read_csv("MaxNoSkusEffect.csv")
dbEffectNoWorkers = pd.read_csv("NoWorkersEffect.csv")
dbEffectRangeCols = pd.read_csv("RangeColRowsEffect.csv")
```

```python
dbEffectNoSkus.describe()
```

```python
"""
Plot time taken against no of SKUS
"""
plt.scatter(dbEffectNoSkus["NO SKUS"],dbEffectNoSkus["TIME TAKEN RANDOM"])
plt.scatter(dbEffectNoSkus["NO SKUS"],dbEffectNoSkus["TIME TAKEN HEURISTIC"])
plt.scatter(dbEffectNoSkus["NO SKUS"],dbEffectNoSkus["TIME TAKEN OPTIMAL"])

plt.xlabel("No Skus")
plt.ylabel("Time Taken (seconds)")
plt.legend(["Random","Heuristic","Optimal"])
plt.show()

```

```python
"""
Plot distance travelled against no of SKUS
"""
plt.scatter(dbEffectNoSkus["NO SKUS"],dbEffectNoSkus["DIST TRAVELLED RANDOM"])
plt.scatter(dbEffectNoSkus["NO SKUS"],dbEffectNoSkus["DIST TRAVELLED HEURISTIC"])
plt.scatter(dbEffectNoSkus["NO SKUS"],dbEffectNoSkus["DIST TRAVELLED OPTIMAL"])

plt.xlabel("No Skus")
plt.ylabel("Total Distance Travelled (meters)")
plt.legend(["Random","Heuristic","Optimal"])
plt.show()
```

```python
"""
Plot time taken against no of SKUS
"""
noPickersData = dbEffectNoWorkers.groupby("NO OF PICKERS", as_index = False)
dataRandom = noPickersData["TIME TAKEN RANDOM"].mean()["TIME TAKEN RANDOM"]
dataHeuristic = noPickersData["TIME TAKEN HEURISTIC"].mean()["TIME TAKEN HEURISTIC"]
dataOptimal = noPickersData["TIME TAKEN OPTIMAL"].mean()["TIME TAKEN OPTIMAL"]
x = np.arange(4,step=1)
width = 0.20

plt.bar(x-0.2,dataRandom,width)
plt.bar(x,dataHeuristic,width)
plt.bar(x+0.2,dataOptimal,width)

plt.xlabel("No Workers")
plt.ylabel("Time Taken (seconds)")
plt.xticks(x)
plt.legend(["Random","Heuristic","Optimal"])

plt.show()
```

```python
"""
Plot distance travelled against no of SKUS
"""
noPickersData = dbEffectNoWorkers.groupby("NO OF PICKERS", as_index = False)
dataRandom = noPickersData["DIST TRAVELLED RANDOM"].mean()["DIST TRAVELLED RANDOM"]
dataHeuristic = noPickersData["DIST TRAVELLED HEURISTIC"].mean()["DIST TRAVELLED HEURISTIC"]
dataOptimal = noPickersData["DIST TRAVELLED OPTIMAL"].mean()["DIST TRAVELLED OPTIMAL"]
x = np.arange(1,5,step=1)
width = 0.20

plt.bar(x-0.2,dataRandom,width)
plt.bar(x,dataHeuristic,width)
plt.bar(x+0.2,dataOptimal,width)

plt.xlabel("No Workers")
plt.ylabel("Distance Travelled (meters)")
plt.xticks(x)
plt.legend(["Random","Heuristic","Optimal"])

plt.show()
```

```python
"""
Plot time taken against no of SKUS
"""
rowsColsData = dbEffectRangeCols.groupby("NO ROWS COLS", as_index = False)
dataRandom = rowsColsData["TIME TAKEN RANDOM"].mean()["TIME TAKEN RANDOM"]
dataHeuristic = rowsColsData["TIME TAKEN HEURISTIC"].mean()["TIME TAKEN HEURISTIC"]
dataOptimal = rowsColsData["TIME TAKEN OPTIMAL"].mean()["TIME TAKEN OPTIMAL"]
x = np.arange(1,10,step=1)
width = 0.20

plt.bar(x-0.2,dataRandom,width)
plt.bar(x,dataHeuristic,width)
plt.bar(x+0.2,dataOptimal,width)

plt.xlabel("No Rows Cols")
plt.ylabel("Time Taken (seconds)")
plt.xticks(x)
plt.legend(["Random","Heuristic","Optimal"])

plt.show()
```

```python
"""
Plot distance travelled against no of SKUS
"""
rowsColsData = dbEffectRangeCols.groupby("NO ROWS COLS", as_index = False)
dataRandom = rowsColsData["DIST TRAVELLED RANDOM"].mean()["DIST TRAVELLED RANDOM"]
dataHeuristic = rowsColsData["DIST TRAVELLED HEURISTIC"].mean()["DIST TRAVELLED HEURISTIC"]
dataOptimal = rowsColsData["DIST TRAVELLED OPTIMAL"].mean()["DIST TRAVELLED OPTIMAL"]
x = np.arange(1,10,step=1)
width = 0.20

plt.bar(x-0.2,dataRandom,width)
plt.bar(x,dataHeuristic,width)
plt.bar(x+0.2,dataOptimal,width)

plt.xlabel("No Rows Cols")
plt.ylabel("Distance Travelled (meters)")
plt.xticks(x)
plt.legend(["Random","Heuristic","Optimal"])

plt.show()
```
